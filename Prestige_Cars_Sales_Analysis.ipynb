```python
# استيراد المكتبات المطلوبة
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots as sp
from plotly.subplots import make_subplots

# تحميل البيانات
df = pd.read_excel('/content/Prestige Cars Data.xlsx')
```

```python
# عرض معلومات أساسية
print(f"أبعاد البيانات: {df.shape}")
print("\nعينة من البيانات:")
display(df.head())

print("\nمعلومات البيانات:")
display(df.info())

```

```python
print("\nالقيم المفقودة:")
display(df.isnull().sum())
```

```python

num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

cat_cols = df.select_dtypes(exclude=[np.number]).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("\nالقيم المفقودة:")
display(df.isnull().sum())
```

```python
import pandas as pd
import plotly.express as px




# تحويل التواريخ
df['SaleDate'] = pd.to_datetime(df['SaleDate'])
df['Stock.DateBought'] = pd.to_datetime(df['Stock.DateBought'])
df['MakeID'] = df['MakeID'].astype(int)
# 2. استخراج اسم الشهر ورقم الشهر (عشان الترتيب)
df['SaleMonthName'] = df['SaleDate'].dt.month_name()
df['SaleMonth'] = df['SaleDate'].dt.month

# 3. التجميع حسب الشهر (بغض النظر عن السنة)
monthly_profit = df.groupby(['SaleMonth', 'SaleMonthName'])['PROFIT'].sum().reset_index()

# 4. ترتيب البيانات حسب رقم الشهر (1, 2, 3...)
monthly_profit = monthly_profit.sort_values('SaleMonth')

# إضافة أعمدة الشهر والسنة للتحليل الزمني
df['SaleYearMonth'] = df['SaleDate'].dt.to_period('M').astype(str)


# 3. إنشاء الرسومات البيانية (Visualizations)

# الرسم الأول: إجمالي المبيعات حسب الدولة (Bar Chart)
country_sales = df.groupby('CountryName')['TotalSalePrice'].sum().reset_index().sort_values(by='TotalSalePrice', ascending=False)
fig1 = px.bar(country_sales,
              x='CountryName',
              y='TotalSalePrice',
              title='Total Sales by Country (إجمالي المبيعات حسب الدولة)',
              color='TotalSalePrice',
              labels={'TotalSalePrice': 'Sales Amount (£)', 'CountryName': 'Country'})
fig1.show()

# 5. الرسم البياني
fig2 = px.line(monthly_profit,
               x='SaleMonthName',
               y='PROFIT',
               title='Monthly Profit Seasonality (إجمالي الأرباح المجمعة لكل شهر)',
               markers=True,
               labels={'PROFIT': 'Total Profit (£)', 'SaleMonthName': 'Month'})
fig2.show()

# الرسم الثالث: الأرباح حسب نوع السيارة (Bar Chart)
make_profit = df.groupby('MakeName')['PROFIT'].sum().reset_index().sort_values(by='PROFIT', ascending=False)
fig3 = px.bar(make_profit,
              x='MakeName',
              y='PROFIT',
              title='Total Profit by Car Make (الأرباح حسب نوع السيارة)',
              color='PROFIT',
              color_continuous_scale='Viridis',
              labels={'PROFIT': 'Profit (£)', 'MakeName': 'Car Make'})
fig3.show()


# الرسم الرابع: علاقة سعر البيع بالتكلفة (Scatter Plot)
# حجم الدائرة يعبر عن سعر البيع، واللون يعبر عن نوع السيارة
fig4 = px.scatter(df,
                  x='Total Cost',
                  y='TotalSalePrice',
                  color='MakeName',
                  size='TotalSalePrice',
                  hover_data=['ModelName', 'PROFIT'],
                  title='Sale Price vs Total Cost (سعر البيع مقابل التكلفة)',
                  labels={'Total Cost': 'Total Cost (£)', 'TotalSalePrice': 'Sale Price (£)'})
fig4.show()
```
